Part - I Idetification and Clearning Missing Values

```python
#Drop the column area_water_percent
df.drop(columns=['area_water_percent'],inplace=True)
df
```

```python
#Drop the rows having missing values and store your resulting DataFrame in the variable
df_narows_dropped = df.dropna()
df_narows_dropped
```

```python
#Drop Rows with More Than 5 Missing Values
df_rows_dropped = df.dropna(thresh=5,axis=0)
df_rows_dropped
```

```python
#Remove Columns with Missing Values
df_nacols_dropped = df.dropna(axis=1)
df_nacols_dropped
```

```python
#Drop colomns with more than 10 missing values and store the resulting DataFrame in the variable df_cols_dropped
df_cols_dropped = df.dropna(axis=1, thresh=len(df) - 10)
df_cols_dropped
```

Part - II Data Imputation Methods

```python
#Fill the missing values in elevation_m with -999. Store your result in the variable filled_elevation_m
filled_elevation_m = df['elevation_m'].fillna(-999)
filled_elevation_m
```

```python
#Fill the missing values in area_total_km2 with the value 0 permanently, store your result in the variable filled_area_total
filled_area_total = df['area_total_km2'].fillna(0,inplace=True)
```

```python
#Fill the missing values of the column latd using backward filling method and store your result in the variable bfill_latd
bfill_latd = df['latd'].bfill()
bfill_latd
```

```python
#Fill the missing values of the column longd using forwarding filling method and store your result in the variable ffill_longd
#limit=15 parameter is used to limit the number of consecutive missing values to be filled
ffill_longd = df['longd'].ffill()
ffill_longd
```

```python
#Fill the missing values of the column population_total with the mean of the column and store your result in the variable mean_total_population
mean_total_population = df['population_total'].fillna(df['population_total'].mean())
mean_total_population
```

```python
#Fill the missing values of the column area_water_sq_mi with the median value of the column and store your result in the variable median_fill
median_fill = df['area_water_sq_mi'].fillna(df['area_water_sq_mi'].median())
median_fill
```

```python
df['area_land_km2'].mode()
```

```python
#Fill the missing values of the column area_land_km2 with the mode value of the column and store your result in the variable mode_fill
#mode can have 2 values so we use the first
mode_fill = df['area_land_km2'].fillna(df['area_land_km2'].mode()[0])
mode_fill
```

```python
#Write following code will fill the missing values in the DataFrame with zeros and store your result in a variable filled_df
filled_df = df.fillna(0)
filled_df
```

```python
#Which of the following is/are the general structure for filling a column values with any value (mean, mode, or median)
df['column_name'].fillna(df['column_name'].mean())
df['column_name'].fillna(df['column_name'].mode()[0])
df['column_name'].fillna(df['column_name'].median())
```
